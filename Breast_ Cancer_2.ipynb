##### 4. Spliting data into Variables and Target

```python
x = breast_cancer_dataset.drop(columns= ['diagnosis', 'Unnamed: 32'], axis= 1) # Variables dataset
y = breast_cancer_dataset['diagnosis'] # Target dataset
```

```python
x.head()
```

```python
y.head()
```

##### 5. Spilt dataset into train and test data

```python
# Importing libraries
from sklearn.model_selection import train_test_split  
from sklearn.preprocessing import StandardScaler
```

```python
x_train,x_test,y_train, y_test = train_test_split(x ,y, test_size= 0.2, random_state= 2) # 20% of the dataset was used for testing data
```

```python
x_train.shape,x_test.shape,y_train.shape, y_test.shape
# x_train --> training data for the variables
# x_test --> testing data for the variables
# y_train --> training data for the target
# y_test --> testing data for the target
```

```python
scaler = StandardScaler() 
```

```python
x_train_std = scaler.fit_transform(x_train)
x_test_std = scaler.transform(x_test)
```

```python
x_train_std
```

##### Model Training with Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
logistic_model = LogisticRegression()
logistic_model.fit(x_train_std,y_train)
```

```python
prediction_train_data = logistic_model.predict(x_train_std)
prediction_test_data = logistic_model.predict(x_test_std)
```

Model Evaluation

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import f1_score
```

```python
train_data_accuracy = accuracy_score(prediction_train_data, y_train)
print(f'The accuracy score of training data is {train_data_accuracy}')
```

```python
test_data_accuracy = accuracy_score(prediction_test_data, y_test)
print(f'The accuracy score of test data is {test_data_accuracy}')
```

```python
train_data_accuracy= precision_score(prediction_train_data, y_train)
print(f'The precision score of training data is {train_data_accuracy}')
```

```python
test_data_accuracy = precision_score(prediction_test_data, y_test)
print(f'The precision score of testing data is {test_data_accuracy}')
```

```python
train_data_accuracy= f1_score(prediction_train_data, y_train)
print(f'The f1 score of training data is {train_data_accuracy}')
```

```python
test_data_accuracy = f1_score(prediction_test_data, y_test)
print(f'The f1 score of testing data is {test_data_accuracy}')
```

Model Optimization

```python
from sklearn.model_selection import GridSearchCV
```

```python
# Define hyperparameter grid
param_grid = {
    'C': [0.01, 0.1, 1, 10, 100],  
    'penalty': ['l1', 'l2'],        
    'solver': ['liblinear', 'saga'] 
}
```

```python
# Initialize GridSearchCV
grid_search = GridSearchCV(estimator= logistic_model, param_grid=param_grid, cv=5, verbose=2, n_jobs=-1)
```

```python
#Fit grid search CV to data
grid_search.fit(x_train_std, y_train)
```

```python
# Get best parameters
best_params = grid_search.best_params_
```

```python
# Use the best model
best_model= grid_search.best_estimator_
```

```python
# Evaluate teh model
test_accuracy = best_model.score(x_test_std,y_test)
print(f'The accuracy score of the testing dataset is {test_accuracy}')
```

```python
from sklearn.metrics import classification_report
```

```python
test_prediction = best_model.predict(x_test_std)
report = classification_report(test_prediction, y_test)
print(f'The classification report:\n {report}')
```
